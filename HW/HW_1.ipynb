## 5. Развернуть число - 0.5 балл

Вам на вход подается произвольное целое число. "Разверните" его, не используя преобразование к строке.

Hint: нужно использовать операторы деления для получения остатка и целого числа, что мы с вами проходили.

Пример:
    
    Вход: 12345890
    Выход: 9854321

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# will the code be with you
# solution here

n=int(input())
number=0
while n>0:
    number=number*10+n%10
    n=n//10
print(number)
#принимаю на входе число, и пока оно больше нуля делаю цикл. в переменную умножаю на 10 
#и прибавляю остаток от деления введенного числа на 10 (таким образом как бы сдвигаю число влево, чтоб вставить новую цифру),
#для самого числа делаю целочисленное деление на 10 (чтоб как бы избавиться от цифры, которое перед эти добавила в переменную)
```

## 6. Азбука Морзе - (0.75 балла)
 
Напишите программу, которая по введённому английскому тексту кодирует каждое слово в [азбуке Морзе](https://en.wikipedia.org/wiki/Morse_code).

Каждая буква заменяется на последовательность точек и тире. В качестве тире используйте обычный дефис `-`, а в качестве точки — точку `.`.  

Например, буква **B** превращается в `-...`. Между закодированными буквами ставится ровно один пробел.  Например, слово Help превратится в `.... . .-.. .--..`. Обратите внимание, что строчные и заглавные буквы кодируются одинаково.

**Формат ввода**
Весь текст записан в единственной строке. Текст состоит из английских букв и пробелов, других символов в тексте нет. В тексте не может быть двух или более пробелов подряд. 

**Формат вывода**
Выведите каждое слово исходного текста, закодированное азбукой Морзе, на отдельной строке. Количество строк в ответе должно совпадать с количеством слов в исходном тексте.

**Пример:**

Ввод: 

Help me SOS

Вывод:

.... . .-.. .--.  
-- .  
... --- ...  



```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# will the code be with you
# solution here

morse={'a':'.-','b':'-...', 'c':'-.-.', 'd':'-..',
       'e':'.','f':'..-.', 'g':'--.', 'h':'....',
       'i':'..','j':'.---', 'k':'-.-', 'l':'.-..',
       'm':'--','n':'-.', 'o':'---', 'p':'.--.',
       'q':'--.-','r':'.-.', 's':'...', 't':'-',
       'u':'..-','v':'...-', 'w':'.--', 'x':'-..-',
       'y':'-.--','z':'--..',}
t=input().lower()
words=t.split()
for word in words:
    m=[]
    for i in word:
        m.append(morse[i])
    print(' '.join(m))
#сделала словарь с азбукой морзе, введенный текст делаю в lower case, чтоб не мучаться с регистром
#разделила текст на слова, для каждого слова создаю пустой список
#для каждой буквы в слове добавляю в пустой список ее кодировку по азбуке морзе и потом вывожу эти списки, где буквы выводятся через пробел 
```

## 7. Классный прямоугольник (0.5 балла)

Разработайте класс `Rectangle`. При инициализации класс принимает два кортежа рациональных координат противоположных углов прямоугольника (со сторонами, параллельными осям координат).  

Класс должен реализовывать методы:
- `perimeter` — возвращает периметр прямоугольника;
- `area` — возвращает площадь прямоугольника.

Все результаты вычислений нужно округлить до сотых.

### Примечание

Ваше решение должно содержать только классы и функции.  

---

### Пример

**Ввод**

rect = Rectangle((3.2, -4.3), (7.52, 3.14))
print(rect.perimeter())

**Вывод**

23.52

**Ввод**

rect = Rectangle((7.52, -4.3), (3.2, 3.14))
print(rect.area())

**Вывод**

32.14


```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# will the code be with you
# solution here

class Rectangle:
    def __init__(self, p1, p2):
        self.l=abs(p2[0]-p1[0])
        self.w=abs(p2[1]-p1[1])

    def perimeter(self):
        return round((self.l+self.w)*2,2)
    
    def area(self):
        return round(self.l*self.w,2)

rect = Rectangle((3.2, -4.3), (7.52, 3.14)) 
print(rect.perimeter())

rect = Rectangle((7.52, -4.3), (3.2, 3.14))
print(rect.area())
#создала класс, определила стороны прямоугольника, вычислив модуль разности координат точек (модуль тут нужен, тк не только натуральные
#числа, а иначе иногда стороны бы неправильно посчитались)
#написала функции расчета периметра и площади, округленных до сотых и проверила все
```

## 8. Объединение словарей (0.75 балла)

Напишите программу, которая подсчитывает значения из произвольного списка словарей следующего вида.

Ввод:

    shops = [{'товар': 'яблоки', 'количество': 400}, {'товар': 'конфеты', 'количество': 300}, {'товар': 'яблоки', 'количество': 750}]


Вывод:

    {'яблоки': 1150, 'конфеты': 300}

```python
shops = [{'товар': 'яблоки', 'количество': 400},
         {'товар': 'конфеты', 'количество': 300}, 
         {'товар': 'яблоки', 'количество': 750}]
```

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# will the code be with you
# solution here

d={}
for i in shops:
    if i['товар'] in d:
        d[i['товар']]+=i['количество']
    else:
        d[i['товар']]=i['количество']
print(d)
#создала пустой словарь, куда буду вписывать сложенные словари
#для каждого словаря из списка я проверяю, если товар уже есть в созданном словаре, то тогда прибавляю количество из анализируемого словаря
#если же в созданном словаре еще нет  такого товара, то создается новый ключ с количеством этого товара
```
